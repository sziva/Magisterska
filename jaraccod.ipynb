```python
folder_path = '/Users/Guest/Downloads/Magisterska-main/tINITmodels'
output_path = '/Users/Guest/Downloads/Magisterska-main/tINITmodels/jarac'
```

```python
filenames = os.listdir(folder_path)
filenames.sort()
```

```python
result_data = []

for i in range(1, len(filenames), 2):  # Iterate with a step of 2 to get consecutive pairs

        file_control = filenames[i]
        file_kd = filenames[i + 1]
        
        print(file_control)
        print(file_kd)
        
        file_path_c = os.path.join(folder_path, file_control)
        file_path_kf = os.path.join(folder_path, file_kd)

        model_control = load_matlab_model(file_path_c)
        model_kd = load_matlab_model(file_path_kf)
        
        reacts_control = {r.id for r in model_control.reactions}
        reacts_kd = {r.id for r in model_kd.reactions}
        
        J = len(reacts_control & reacts_kd)/len(reacts_control | reacts_kd)
        print(J)
        
        result_data.append({'File_Control': file_control, 'J_Value': J})
        print(result_data)
        
# Create a DataFrame from the list of dictionaries
result_data_df = pd.DataFrame(result_data)

# Save the DataFrame to an Excel file
result_excel_file = 'J_values.xlsx'
result_data_df.to_excel(result_excel_file, index=False)
print(f'J values saved to {result_excel_file}')
```
